```python
import pandas as pd

x = pd.read_csv("sales_data_sample.csv", encoding='latin1')
print(x.head())
```

```python
x.describe()
```

```python
print(x.isnull().sum())
```

```python
# Identify and handle missing values using .isnull() in Python or filters in Excel.

x = x.drop(columns = ['ADDRESSLINE2', 'STATE', 'TERRITORY'])
x = x.dropna(subset=['ORDERLINENUMBER'])
numerical_cols=['QUANTITYORDERED','SALES']
for col in numerical_cols:
  # Calculate the mean of non-missing values in the column
  col_mean = x[col].mean()

  # Fill missing values in the column with the calculated mean
  x[col] = x[col].fillna(col_mean)
categorical_cols =['PRODUCTLINE','POSTALCODE']

for col in categorical_cols:
  # Fill missing values in the column with the forward fill
  x[col] = x[col].ffill()

print(x.head())
```

```python
print(x.isnull().sum())
```

```python
# Remove duplicate rows using .drop_duplicates() or Excel’s “Remove Duplicates”.

print(f"Data shape before removing duplicates: {x.shape}")
num_duplicates = x.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")
x.drop_duplicates(inplace=True)
print(f"Data shape after removing duplicates: {x.shape}")
```

```python
# Standardize text values like gender, country names, etc.

print(x[['COUNTRY', 'CITY', 'STATUS', 'CUSTOMERNAME']].head())
x['CUSTOMERNAME'] = x['CUSTOMERNAME'].astype(str).str.strip().str.title()
x['COUNTRY'] = x['COUNTRY'].str.upper()
x['STATUS'] = x['STATUS'].str.upper()
x['COUNTRY'] = x['COUNTRY'].replace({
    'United States': 'USA',
    'U.S.A.': 'USA',
    'Usa': 'USA',
    'Uk': 'UK',
    'United Kingdom': 'UK'
})
print(x[['COUNTRY', 'CITY', 'STATUS', 'CUSTOMERNAME']].head())
```

```python
# Convert date formats to a consistent type (e.g., dd-mm-yyyy).

print("Before standardizing:")
print(x['ORDERDATE'].head())
x['ORDERDATE'] = x['ORDERDATE'].str.replace('-', '/', regex=False)

# Convert using explicit format: MM/DD/YYYY H:M
x['ORDERDATE'] = pd.to_datetime(x['ORDERDATE'], format='%m/%d/%Y %H:%M', errors='coerce')

# Format to dd-mm-yyyy
x['ORDERDATE'] = x['ORDERDATE'].dt.strftime('%d-%m-%Y')

print("\nAfter standardizing:")
x['ORDERDATE'].head()
```

```python
# Rename column headers to be clean and uniform (e.g., lowercase, no spaces).
x.columns = x.columns.str.strip().str.lower().str.replace(' ', '_')
print(x.columns)
```

```python
# Check and fix data types (e.g., age should be int, date as datetime).

print('Before:')
print(x.dtypes)
x['orderdate'] = pd.to_datetime(x['orderdate'], errors='coerce', format='%d-%m-%Y')
x['quantityordered'] = x['quantityordered'].round().astype('Int64')
x['orderlinenumber'] = x['orderlinenumber'].astype('Int64')
print('After:')
print(x.dtypes)
```

```python
x.to_csv('cleaned_sales_data_sample.csv', index=False)
from google.colab import files
files.download('cleaned_sales_data_sample.csv')
```
